```
# for duc 2001 our approach
foname = np.load(common_path + 'folder_name_2001' + '.npy')
ic = np.load(common_path + 'colecciones_testing_pyrouge.npy')
foname_new = []
for ele in ic:
  foname_new.append(foname[ele])
```

```
###### delete sentences in system summary for DUC 2001 Our approach balanced MLP-relu pyrouge
common_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/'
import os

for i in range(len(foname_new)):

  # for DUC 2001 Our approach unbalanced MLP-relu pyrouge
  goldfile = open(common_path + 'DUC_2001_training_gold_standard_filtered/' + '{}_englishReference1.txt'.format(foname_new[i]), 'r').readlines()
 
  # for DUC 2001
  num_cl = 4
  infile = open(common_path + 'testing_for_pltrdy_classifier_pyrouge_{}_0/'.format(num_cl) + 'GaussianNB_200_LEMA_{}/'.format(0) + '{}_englishSyssum1.txt'.format(foname_new[i]), 'r').readlines()

  #print(i, len(goldfile), len(infile))

  # for DUC 2001
  folder = 'summaries_cl{}_for_ROUGE_pltrdy_DUC2001_pyrouge/'.format(num_cl)
  if os.path.exists(common_path + folder) == False:
    os.mkdir(common_path + folder)

  file_name = '{}_englishSyssum1.txt'.format(foname_new[i])

  with open(common_path + folder + file_name, 'w') as outfile:
    for index, line in enumerate(infile):
      if index < len(goldfile):
        outfile.write(line)
```

```
# duc 2001 our approach mlp-relu balanced data
common_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/'
scores = []
aux_rouge1_r = 0
aux_rouge1_p = 0
aux_rouge1_f = 0
aux_rouge2_r = 0
aux_rouge2_p = 0
aux_rouge2_f = 0
aux2_rouge1_r = []
aux2_rouge1_p = []
aux2_rouge1_f = []
aux2_rouge2_r = []
aux2_rouge2_p = []
aux2_rouge2_f = []

num_cl = 4

for j in range(len(foname_new)): # for DUC 2001

  # print(j)

  # for DUC 2001 A pyrouge
  folder = 'summaries_cl{}_for_ROUGE_pltrdy_DUC2001_pyrouge/'.format(num_cl)
  hyp_path = common_path + folder + '{}_englishSyssum1.txt'.format(foname_new[j])
  ref_path = common_path + 'DUC_2001_training_gold_standard_filtered/' + '{}_englishReference1.txt'.format(foname_new[j])


  from rouge import FilesRouge

  files_rouge = FilesRouge(hyp_path, ref_path)
  scores = files_rouge.get_scores()
  # or
  # scores = files_rouge.get_scores(avg=True)

  for i in range(len(scores)):
    # print(i)
    aux_rouge1_r += scores[i]['rouge-1']['r'] #ROUGE-1 recall
    aux_rouge1_p += scores[i]['rouge-1']['p'] #ROUGE-1 precision
    aux_rouge1_f += scores[i]['rouge-1']['f'] #ROUGE-1 F1 score
    aux_rouge2_r += scores[i]['rouge-2']['r'] #ROUGE-2 recall
    aux_rouge2_p += scores[i]['rouge-2']['p'] #ROUGE-2 precision
    aux_rouge2_f += scores[i]['rouge-2']['f'] #ROUGE-2 F1 score

  # print(j, '{:5.4f}'.format(aux_rouge1/len(scores)))
  # print('{:5.4f}'.format(aux_rouge2/len(scores)))
  # print('{:5.4f}'.format(aux_fscore/len(scores)))

  aux2_rouge1_r.append(aux_rouge1_r/len(scores))
  aux2_rouge1_p.append(aux_rouge1_p/len(scores))
  aux2_rouge1_f.append(aux_rouge1_f/len(scores))
  aux2_rouge2_r.append(aux_rouge2_r/len(scores))
  aux2_rouge2_p.append(aux_rouge2_p/len(scores))
  aux2_rouge2_f.append(aux_rouge2_f/len(scores))

  aux_rouge1_r = 0
  aux_rouge1_p = 0
  aux_rouge1_f = 0
  aux_rouge2_r = 0
  aux_rouge2_p = 0
  aux_rouge2_f = 0

foname = foname_new

print('ROUGE-1 recall = {:5.4f}'.format(sum(aux2_rouge1_r)/len(foname)))
print('ROUGE-1 precision = {:5.4f}'.format(sum(aux2_rouge1_p)/len(foname)))
print('ROUGE-1 F1 score = {:5.4f}'.format(sum(aux2_rouge1_f)/len(foname)))
print('ROUGE-2 recall = {:5.4f}'.format(sum(aux2_rouge2_r)/len(foname)))
print('ROUGE-2 precision = {:5.4f}'.format(sum(aux2_rouge2_p)/len(foname)))
print('ROUGE-2 F1 score = {:5.4f}'.format(sum(aux2_rouge2_f)/len(foname)))

np.save(common_path + 'ROUGE1_recall_DUC2001_training_sup_cl{}_pyrouge'.format(num_cl) + '.npy', sum(aux2_rouge1_r)/len(foname))
np.save(common_path + 'ROUGE1_F1score_DUC2001_training_sup_cl{}_pyrouge'.format(num_cl) + '.npy', sum(aux2_rouge1_f)/len(foname))
np.save(common_path + 'ROUGE1_precision_DUC2001_training_sup_cl{}_pyrouge'.format(num_cl) + '.npy', sum(aux2_rouge1_p)/len(foname))
np.save(common_path + 'ROUGE2_recall_DUC2001_training_sup_cl{}_pyrouge'.format(num_cl) + '.npy', sum(aux2_rouge2_r)/len(foname))
np.save(common_path + 'ROUGE2_precision_DUC2001_training_sup_cl{}_pyrouge'.format(num_cl) + '.npy', sum(aux2_rouge2_p)/len(foname))
np.save(common_path + 'ROUGE2_F1score_DUC2001_training_sup_cl{}_pyrouge'.format(num_cl) + '.npy', sum(aux2_rouge2_f)/len(foname))
```
