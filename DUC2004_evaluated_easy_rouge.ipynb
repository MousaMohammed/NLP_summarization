```
def summary_evaluation_ROUGEN(docs_per_sentence, gold_per_sentence, n_rouge):

  from rouge.rouge import rouge_n_sentence_level

  def prepare_results(r, p, f, n_rouge):
      return '\t{}:{:5.4f}\t{}: {:5.4f}\t{}: {:5.4f}'.format('ROUGE{}-recall'.format(n_rouge), r, 'ROUGE{}-precision'.format(n_rouge), p, 'ROUGE{}-F1 score'.format(n_rouge), f)

  j = 0
  RN_recall_total = []
  RN_precision_total = []
  RN_fscore_total = []
  
  for j in range(len(docs_per_sentence)):
      
    RN_recall, RN_precision, RN_fscore = rouge_n_sentence_level(docs_per_sentence[j], gold_per_sentence[j], n_rouge, 1)
    RN_recall_total.append(RN_recall)
    RN_precision_total.append(RN_precision)
    RN_fscore_total.append(RN_fscore)
  
  print(prepare_results(RN_recall, RN_precision, RN_fscore, n_rouge))
  
  return RN_recall_total, RN_precision_total, RN_fscore_total
```

```
import numpy as np
common_path = "/content/gdrive/My Drive/Codigos_doctorado_DUC2004/"
ic = np.load(common_path + 'colecciones_testing_DUC2004_pyrouge.npy')
foname = np.load(common_path + 'folder_name_2004' + '.npy')
sent_per_collection = np.load(common_path + 'summaries_per_collection_DUC2004_balanced' + '.npy')
files_names = np.load('/content/gdrive/My Drive/Codigos_doctorado_DUC2001/' + 'files_names_longitude_DUC2004' + '.npy')
gold = np.load('/content/gdrive/My Drive/Codigos_doctorado_DUC2001/' + 'documentos_por_palabras_gold_testing_DUC2004' + '.npy')
```

```
golda = []
goldb = []
goldc = []
goldd = []

aux_files_a = []
aux_files_b = []
aux_files_c = []
aux_files_d = []

for i in range(len(files_names)):
  if i%4 == 0:
    aux_files_a.append(files_names[i][0:4])
    golda.append(gold[i])

for i in range(len(files_names)):
  if i%4 == 1:
    aux_files_b.append(files_names[i][0:4])
    goldb.append(gold[i])

for i in range(len(files_names)):
  if i%4 == 2:
    aux_files_c.append(files_names[i][0:4])
    goldc.append(gold[i])

for i in range(len(files_names)):
  if i%4 == 3:
    aux_files_d.append(files_names[i][0:4])
    goldd.append(gold[i])
```
