```
###### delete sentences in system summary for DUC 2002 CluAggloRoulD unsup (30 experiments)
common_path = '/content/gdrive/My Drive/Codigos_doctorado_unsupervised_best/'
import os

for j in range(30):
  for i in range(len(foname)):

    # for DUC 2002 CluAggloRoulD (30 experiments)
    goldfile = open('/content/gdrive/My Drive/Codigos_doctorado_DUC2001/' + 'gold_standardA_DUC2002/' + '{}_englishReference1.txt'.format(foname[i]), 'r').readlines()
    # goldfile = open('/content/gdrive/My Drive/Codigos_doctorado_DUC2001/' + 'gold_standardB_DUC2002/' + '{}_englishReference1.txt'.format(foname[i]), 'r').readlines()

    # for DUC 2002
    infile = open(common_path + 'CluAgglo_RoulD_DUC2002_resumenes2/' + 'CluAgglo_RoulD_resumenes_{}/'.format(j) + '{}_englishSyssum1.txt'.format(foname[i]), 'r').readlines()
    
    # for DUC 2002
    folder = 'summaries_for_ROUGE_pltrdy_A_DUC2002/'
    if os.path.exists(common_path + folder) == False:
      os.mkdir(common_path + folder)

    folder2 = 'CluAgglo_RoulD_resumenes_{}/'.format(j)
    if os.path.exists(common_path + folder + folder2) == False:
      os.mkdir(common_path + folder + folder2)

    # print(i, len(goldfile), len(infile))

    file_name = '{}_englishSyssum1.txt'.format(foname[i])

    with open(common_path + folder + folder2 + file_name, 'w') as outfile:
      for index, line in enumerate(infile):
        if index < len(goldfile):
          outfile.write(line)
```

```
# for DUC 2002 unsupervised (30 experiments)
common_path = '/content/gdrive/My Drive/Codigos_doctorado_unsupervised_best/'
scores = []
aux_rouge1_r = 0
aux_rouge1_p = 0
aux_rouge1_f = 0
aux_rouge2_r = 0
aux_rouge2_p = 0
aux_rouge2_f = 0

aux2_rouge1_r = []
aux2_rouge1_p = []
aux2_rouge1_f = []
aux2_rouge2_r = []
aux2_rouge2_p = []
aux2_rouge2_f = []

aux3_rouge1_r = 0
aux3_rouge1_p = 0
aux3_rouge1_f = 0
aux3_rouge2_r = 0
aux3_rouge2_p = 0
aux3_rouge2_f = 0

for z in range(30):
# for z in range(1):

  for j in range(len(foname)):
  # for j in range(1):

    # print(j)
    # for DUC 2002 A
    # hyp_path = common_path + 'summaries_for_ROUGE_pltrdy_A_DUC2002/' + 'CluAgglo_RoulD_resumenes_{}/'.format(z) + '{}_englishSyssum1.txt'.format(foname[j])
    # ref_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/' + 'gold_standardA_DUC2002/' + '{}_englishReference1.txt'.format(foname[j])

    # for DUC 2002 B
    hyp_path = common_path + 'summaries_for_ROUGE_pltrdy_B_DUC2002/' + 'CluAgglo_RoulD_resumenes_{}/'.format(z) + '{}_englishSyssum1.txt'.format(foname[j])
    ref_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/' + 'gold_standardB_DUC2002/' + '{}_englishReference1.txt'.format(foname[j])
    
    from rouge import FilesRouge

    files_rouge = FilesRouge(hyp_path, ref_path)
    scores = files_rouge.get_scores()
    # or
    # scores = files_rouge.get_scores(avg=True)

    for i in range(len(scores)): # average the number of sentences between output summary and gold standard
      # print(i)
      aux_rouge1_r += scores[i]['rouge-1']['r'] #ROUGE-1 recall
      aux_rouge1_p += scores[i]['rouge-1']['p'] #ROUGE-1 precision
      aux_rouge1_f += scores[i]['rouge-1']['f'] #ROUGE-1 F1 score
      aux_rouge2_r += scores[i]['rouge-2']['r'] #ROUGE-2 recall
      aux_rouge2_p += scores[i]['rouge-2']['p'] #ROUGE-2 precision
      aux_rouge2_f += scores[i]['rouge-2']['f'] #ROUGE-2 F1 score

    # print(j, '{:5.4f}'.format(aux_rouge1/len(scores)))
    # print('{:5.4f}'.format(aux_rouge2/len(scores)))
    # print('{:5.4f}'.format(aux_fscore/len(scores)))

    aux2_rouge1_r.append(aux_rouge1_r/len(scores))
    aux2_rouge1_p.append(aux_rouge1_p/len(scores))
    aux2_rouge1_f.append(aux_rouge1_f/len(scores))
    aux2_rouge2_r.append(aux_rouge2_r/len(scores))
    aux2_rouge2_p.append(aux_rouge2_p/len(scores))
    aux2_rouge2_f.append(aux_rouge2_f/len(scores))

    # print(len(aux_rouge1_r))

    aux_rouge1_r = 0
    aux_rouge1_p = 0
    aux_rouge1_f = 0
    aux_rouge2_r = 0
    aux_rouge2_p = 0
    aux_rouge2_f = 0

  # print('Experiment: ', z)
  # print('ROUGE-1 = {:5.4f}'.format(sum(aux2_rouge1)/len(foname)))
  # print('ROUGE-2 = {:5.4f}'.format(sum(aux2_rouge2)/len(foname)))
  # print('F1 score = {:5.4f}'.format(sum(aux2_fscore)/len(foname)))

  # np.save(common_path + 'ROUGE1_recall_DUC2002A_unsup{}'.format(z) + '.npy', sum(aux2_rouge1_r)/len(foname))
  # np.save(common_path + 'ROUGE1_precision_DUC2002A_unsup{}'.format(z) + '.npy', sum(aux2_rouge1_p)/len(foname))
  # np.save(common_path + 'ROUGE1_f1score_DUC2002A_unsup{}'.format(z) + '.npy', sum(aux2_rouge1_f)/len(foname))
  # np.save(common_path + 'ROUGE2_recall_DUC2002A_unsup{}'.format(z) + '.npy', sum(aux2_rouge2_r)/len(foname))
  # np.save(common_path + 'ROUGE2_precision_DUC2002A_unsup{}'.format(z) + '.npy', sum(aux2_rouge2_p)/len(foname))
  # np.save(common_path + 'ROUGE2_f1score_DUC2002A_unsup{}'.format(z) + '.npy', sum(aux2_rouge2_f)/len(foname))

  np.save(common_path + 'ROUGE1_recall_DUC2002B_unsup{}'.format(z) + '.npy', sum(aux2_rouge1_r)/len(foname))
  np.save(common_path + 'ROUGE1_precision_DUC2002B_unsup{}'.format(z) + '.npy', sum(aux2_rouge1_p)/len(foname))
  np.save(common_path + 'ROUGE1_f1score_DUC2002B_unsup{}'.format(z) + '.npy', sum(aux2_rouge1_f)/len(foname))
  np.save(common_path + 'ROUGE2_recall_DUC2002B_unsup{}'.format(z) + '.npy', sum(aux2_rouge2_r)/len(foname))
  np.save(common_path + 'ROUGE2_precision_DUC2002B_unsup{}'.format(z) + '.npy', sum(aux2_rouge2_p)/len(foname))
  np.save(common_path + 'ROUGE2_f1score_DUC2002B_unsup{}'.format(z) + '.npy', sum(aux2_rouge2_f)/len(foname))

  aux3_rouge1_r += sum(aux2_rouge1_r)/len(foname)
  aux3_rouge1_p += sum(aux2_rouge1_p)/len(foname)
  aux3_rouge1_f += sum(aux2_rouge1_f)/len(foname)
  aux3_rouge2_r += sum(aux2_rouge2_r)/len(foname)
  aux3_rouge2_p += sum(aux2_rouge2_p)/len(foname)
  aux3_rouge2_f += sum(aux2_rouge2_f)/len(foname)

  aux2_rouge1_r = []
  aux2_rouge1_p = []
  aux2_rouge1_f = []
  aux2_rouge2_r = []
  aux2_rouge2_p = []
  aux2_rouge2_f = []

# np.save(common_path + 'duc2002_A_unsup_r1_r_rouge_library' + '.npy', aux3_rouge1_r)
# np.save(common_path + 'duc2002_A_unsup_r1_p_rouge_library' + '.npy', aux3_rouge1_p)
# np.save(common_path + 'duc2002_A_unsup_r1_f_rouge_library' + '.npy', aux3_rouge1_f)
# np.save(common_path + 'duc2002_A_unsup_r2_r_rouge_library' + '.npy', aux3_rouge2_r)
# np.save(common_path + 'duc2002_A_unsup_r2_p_rouge_library' + '.npy', aux3_rouge2_p)
# np.save(common_path + 'duc2002_A_unsup_r2_f_rouge_library' + '.npy', aux3_rouge2_f)

np.save(common_path + 'duc2002_B_unsup_r1_r_rouge_library' + '.npy', aux3_rouge1_r)
np.save(common_path + 'duc2002_B_unsup_r1_p_rouge_library' + '.npy', aux3_rouge1_p)
np.save(common_path + 'duc2002_B_unsup_r1_f_rouge_library' + '.npy', aux3_rouge1_f)
np.save(common_path + 'duc2002_B_unsup_r2_r_rouge_library' + '.npy', aux3_rouge2_r)
np.save(common_path + 'duc2002_B_unsup_r2_p_rouge_library' + '.npy', aux3_rouge2_p)
np.save(common_path + 'duc2002_B_unsup_r2_f_rouge_library' + '.npy', aux3_rouge2_f)


print('Average ROUGE1_recall:', aux3_rouge1_r/30)
print('Average ROUGE1_precision:', aux3_rouge1_p/30)
print('Average ROUGE1_f1score:', aux3_rouge1_f/30)
print('Average ROUGE2_recall:', aux3_rouge2_r/30)
print('Average ROUGE2_precision:', aux3_rouge2_p/30)
print('Average ROUGE2_f1score:', aux3_rouge2_f/30)
```
