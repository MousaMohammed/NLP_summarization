```
def summary_evaluation_ROUGEN(docs_per_sentence, gold_per_sentence, n_rouge):

  from rouge.rouge import rouge_n_sentence_level

  def prepare_results(r, p, f, n_rouge):
      return '\t{}:{:5.4f}\t{}: {:5.4f}\t{}: {:5.4f}'.format('ROUGE{}-recall'.format(n_rouge), r, 'ROUGE{}-precision'.format(n_rouge), p, 'ROUGE{}-F1 score'.format(n_rouge), f)

  j = 0
  RN_recall_total = []
  RN_precision_total = []
  RN_fscore_total = []

  for j in range(len(docs_per_sentence)):
      
    RN_recall, RN_precision, RN_fscore = rouge_n_sentence_level(docs_per_sentence[j], gold_per_sentence[j], n_rouge, 1)
    RN_recall_total.append(RN_recall)
    RN_precision_total.append(RN_precision)
    RN_fscore_total.append(RN_fscore)
  
  print(prepare_results(RN_recall, RN_precision, RN_fscore, n_rouge))
  
  return RN_recall_total, RN_precision_total, RN_fscore_total
```

```
import numpy as np
common_path = "/content/gdrive/My Drive/Codigos_doctorado_DUC2001/"
sent_per_collection = np.load(common_path + 'summaries_per_collection_DUC2001_balanced_test_30collectionALL' + '.npy')
hr_golda = np.load(common_path + 'documentos_por_palabras_golda_testing_doubles' + '.npy')
hr_goldb = np.load(common_path + 'documentos_por_palabras_goldb_testing_doubles' + '.npy')
# sac3 = np.load(common_path + 'ext2_sent_num_per_set_2001' + '.npy')
```

```
i = 4
nrouge1 = 1
nrouge2 = 2
gold_std = hr_golda
R1_recall_total_A, R1_precision_total_A, R1_fscore_total_A = summary_evaluation_ROUGEN(sent_per_collection[4][0], gold_std, nrouge1)
R2_recall_total_A, R2_precision_total_A, R2_fscore_total_A = summary_evaluation_ROUGEN(sent_per_collection[4][0], gold_std, nrouge2)

# save results  
np.save(common_path + 'R1_recall_oraciones_DUC2001_classifier_test_30collectionA_{}'.format(i) + '.npy', R1_recall_total_A)
np.save(common_path + 'R1_precision_oraciones_DUC2001_classifier_test_30collectionA_{}'.format(i) + '.npy', R1_precision_total_A)
np.save(common_path + 'R1_fscore_oraciones_DUC2001_classifier_test_30collectionA_{}'.format(i) + '.npy', R1_fscore_total_A)

np.save(common_path + 'R2_recall_oraciones_DUC2001_classifier_test_30collectionA_{}'.format(i) + '.npy', R2_recall_total_A)
np.save(common_path + 'R2_precision_oraciones_DUC2001_classifier_test_30collectionA_{}'.format(i) + '.npy', R2_precision_total_A)
np.save(common_path + 'R2_fscore_oraciones_DUC2001_classifier_test_30collectionA_{}'.format(i) + '.npy', R2_fscore_total_A)
  
```
