```
import numpy as np
common_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/'
# folder name for DUC 2001
foname2 = np.load('/content/gdrive/My Drive/Codigos_doctorado_DUC2001/' + 'files_names_DUC2001' + '.npy') #for gold standards
foname = np.load(common_path + 'folder_name_2001_testing' + '.npy') # for testing collection
```

```
# dividing gold standards for DUC 2001 testing collection
foname_goldA = []
foname_goldB = []

for i in range(len(foname2)):
  aux = i%2
  if aux != 0:
#     print(i)
    foname_goldB.append(foname2[i]) 
  else:
    foname_goldA.append(foname2[i])
```

```
###### delete sentences in system summary for DUC 2001 Original approach balanced MLP-relu
common_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/'
import os

for i in range(len(foname)):

  # for DUC 2001 Original approach balanced MLP-relu
  # goldfile = open(common_path + 'DUC_2001_testing_gold_standard_doubles_filtered/' + '{}_englishReference1.txt'.format(foname_goldA[i]), 'r').readlines()
  goldfile = open(common_path + 'DUC_2001_testing_gold_standard_doubles_filtered/' + '{}_englishReference1.txt'.format(foname_goldB[i]), 'r').readlines()
 
  # for DUC 2001
  num_cl = 4
  infile = open(common_path + 'testing_for_pltrdy_no_batches_classifier_test_30collectionALL_{}/'.format(num_cl) + 'GaussianNB_200_LEMA_{}/'.format(num_cl) + '{}_englishSyssum1.txt'.format(foname[i]), 'r').readlines()

  # print(i, len(goldfile), len(infile))

  # for DUC 2001
  # folder = 'summaries_cl{}_for_ROUGE_pltrdy_A_DUC2001/'.format(num_cl)
  folder = 'summaries_cl{}_for_ROUGE_pltrdy_B_DUC2001/'.format(num_cl)
  if os.path.exists(common_path + folder) == False:
    os.mkdir(common_path + folder)

  file_name = '{}_englishSyssum1.txt'.format(foname[i])

  with open(common_path + folder + file_name, 'w') as outfile:
    for index, line in enumerate(infile):
      if index < len(goldfile):
        outfile.write(line)
```

```
common_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/'
scores = []
aux_rouge1_r = 0
aux_rouge1_p = 0
aux_rouge1_f = 0
aux_rouge2_r = 0
aux_rouge2_p = 0
aux_rouge2_f = 0
aux2_rouge1_r = []
aux2_rouge1_p = []
aux2_rouge1_f = []
aux2_rouge2_r = []
aux2_rouge2_p = []
aux2_rouge2_f = []

num_cl = 3

for j in range(len(foname)):

  # for DUC 2001 A
  folder = 'summaries_cl{}_for_ROUGE_pltrdy_A_DUC2001/'.format(num_cl)
  hyp_path = common_path + folder + '{}_englishSyssum1.txt'.format(foname[j])
  ref_path = common_path + 'DUC_2001_testing_gold_standard_doubles_filtered/' + '{}_englishReference1.txt'.format(foname_goldA[j])

  # for DUC 2001 B
  #folder = 'summaries_cl{}_for_ROUGE_pltrdy_B_DUC2001/'.format(num_cl)
  #hyp_path = common_path + folder + '{}_englishSyssum1.txt'.format(foname[j])
  #ref_path = common_path + 'DUC_2001_testing_gold_standard_doubles_filtered/' + '{}_englishReference1.txt'.format(foname_goldB[j])



  from rouge import FilesRouge

  files_rouge = FilesRouge(hyp_path, ref_path)
  scores = files_rouge.get_scores()
  # or
  # scores = files_rouge.get_scores(avg=True)

  for i in range(len(scores)):
    # print(i)
    aux_rouge1_r += scores[i]['rouge-1']['r'] #ROUGE-1 recall
    aux_rouge1_p += scores[i]['rouge-1']['p'] #ROUGE-1 precision
    aux_rouge1_f += scores[i]['rouge-1']['f'] #ROUGE-1 F1 score
    aux_rouge2_r += scores[i]['rouge-2']['r'] #ROUGE-2 recall
    aux_rouge2_p += scores[i]['rouge-2']['p'] #ROUGE-2 precision
    aux_rouge2_f += scores[i]['rouge-2']['f'] #ROUGE-2 F1 score

  # print(j, '{:5.4f}'.format(aux_rouge1/len(scores)))
  # print('{:5.4f}'.format(aux_rouge2/len(scores)))
  # print('{:5.4f}'.format(aux_fscore/len(scores)))

  aux2_rouge1_r.append(aux_rouge1_r/len(scores))
  aux2_rouge1_p.append(aux_rouge1_p/len(scores))
  aux2_rouge1_f.append(aux_rouge1_f/len(scores))
  aux2_rouge2_r.append(aux_rouge2_r/len(scores))
  aux2_rouge2_p.append(aux_rouge2_p/len(scores))
  aux2_rouge2_f.append(aux_rouge2_f/len(scores))

  aux_rouge1_r = 0
  aux_rouge1_p = 0
  aux_rouge1_f = 0
  aux_rouge2_r = 0
  aux_rouge2_p = 0
  aux_rouge2_f = 0

print('ROUGE-1 recall = {:5.4f}'.format(sum(aux2_rouge1_r)/len(foname)))
print('ROUGE-1 precision = {:5.4f}'.format(sum(aux2_rouge1_p)/len(foname)))
print('ROUGE-1 F1 score = {:5.4f}'.format(sum(aux2_rouge1_f)/len(foname)))
print('ROUGE-2 recall = {:5.4f}'.format(sum(aux2_rouge2_r)/len(foname)))
print('ROUGE-2 precision = {:5.4f}'.format(sum(aux2_rouge2_p)/len(foname)))
print('ROUGE-2 F1 score = {:5.4f}'.format(sum(aux2_rouge2_f)/len(foname)))

np.save(common_path + 'ROUGE1_recall_DUC2001A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge1_r)/len(foname))
np.save(common_path + 'ROUGE1_precision_DUC2001A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge1_p)/len(foname))
np.save(common_path + 'ROUGE1_F1score_DUC2001A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge1_f)/len(foname))
np.save(common_path + 'ROUGE2_recall_DUC2001A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge2_r)/len(foname))
np.save(common_path + 'ROUGE2_precision_DUC2001A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge2_p)/len(foname))
np.save(common_path + 'ROUGE2_F1score_DUC2001A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge2_f)/len(foname))
```
