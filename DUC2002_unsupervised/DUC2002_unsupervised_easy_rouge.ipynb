```
def summary_evaluation_ROUGEN(docs_per_sentence, gold_per_sentence, n_rouge):

  from rouge.rouge import rouge_n_sentence_level

  def prepare_results(r, p, f, n_rouge):
      return '\t{}:{:5.4f}\t{}: {:5.4f}\t{}: {:5.4f}'.format('ROUGE{}-recall'.format(n_rouge), r, 'ROUGE{}-precision'.format(n_rouge), p, 'ROUGE{}-F1 score'.format(n_rouge), f)

  # j = 0
  RN_recall_total = []
  RN_precision_total = []
  RN_fscore_total = []

  for j in range(len(docs_per_sentence)):
      
    RN_recall, RN_precision, RN_fscore = rouge_n_sentence_level(docs_per_sentence[j], gold_per_sentence[j], n_rouge, 1)
    RN_recall_total.append(RN_recall)
    RN_precision_total.append(RN_precision)
    RN_fscore_total.append(RN_fscore)
    
    print(prepare_results(RN_recall, RN_precision, RN_fscore, n_rouge))
  
  return RN_recall_total, RN_precision_total, RN_fscore_total
```

```
# import numpy as np
from nltk import word_tokenize

common_path = '/content/gdrive/My Drive/Codigos_doctorado_unsupervised_best/CluAgglo_RoulD_DUC2002_resumenes2/'
foname = np.load('/content/gdrive/My Drive/Codigos_doctorado/folder_name.npy')
goldA = np.load("/content/gdrive/My Drive/Codigos_doctorado_unsupervised_best/" + 'documentos_por_palabras_goldA_DUC2002' + '.npy')
goldB = np.load("/content/gdrive/My Drive/Codigos_doctorado_unsupervised_best/" + 'documentos_por_palabras_goldB_DUC2002' + '.npy')

docs = []
exp = []
aux = []

for j in range(30):
# for j in range(1):
  for i in range(len(foname)):
  # for i in range(2):
    sent_per_collection = open(common_path + 'CluAgglo_RoulD_resumenes_{}/'.format(j) + '{}_englishSyssum1'.format(foname[i]) + '.txt', 'r').readlines()

    for k in range(len(sent_per_collection)):
      aux.extend(word_tokenize(sent_per_collection[k]))

    docs.append(aux)
    aux = []
  exp.append(docs)
  docs = []
  
```
