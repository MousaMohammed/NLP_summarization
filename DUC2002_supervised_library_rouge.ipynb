```
common_path = '/content/gdrive/My Drive/Codigos_doctorado_DUC2001/'
scores = []
aux_rouge1_r = 0
aux_rouge1_p = 0
aux_rouge1_f = 0
aux_rouge2_r = 0
aux_rouge2_p = 0
aux_rouge2_f = 0
aux2_rouge1_r = []
aux2_rouge1_p = []
aux2_rouge1_f = []
aux2_rouge2_r = []
aux2_rouge2_p = []
aux2_rouge2_f = []

num_cl = 3

for j in range(len(foname_duc2002_sup)): # for DUC 2002

  # print(j)
  # for DUC 2002 A
  hyp_path = common_path + 'summaries_for_ROUGE_pltrdy_A_DUC2002_supervised_{}/'.format(num_cl) + '{}_englishSyssum1.txt'.format(foname_duc2002_sup[j])
  ref_path = common_path + 'model_summariesA_2002_sup/' + '{}_englishReference1.txt'.format(foname_duc2002_sup[j])

  # for DUC 2002 B
  # hyp_path = common_path + 'summaries_for_ROUGE_pltrdy_B_DUC2002_supervised_{}/'.format(num_cl) + '{}_englishSyssum1.txt'.format(foname_duc2002_sup[j])
  # ref_path = common_path + 'model_summariesB_2002_sup/' + '{}_englishReference1.txt'.format(foname_duc2002_sup[j])

  from rouge import FilesRouge

  files_rouge = FilesRouge(hyp_path, ref_path)
  scores = files_rouge.get_scores()
  # or
  # scores = files_rouge.get_scores(avg=True)

  for i in range(len(scores)):
    # print(i)
    aux_rouge1_r += scores[i]['rouge-1']['r'] #ROUGE-1 recall
    aux_rouge1_p += scores[i]['rouge-1']['p'] #ROUGE-1 precision
    aux_rouge1_f += scores[i]['rouge-1']['f'] #ROUGE-1 F1 score
    aux_rouge2_r += scores[i]['rouge-2']['r'] #ROUGE-2 recall
    aux_rouge2_p += scores[i]['rouge-2']['p'] #ROUGE-2 precision
    aux_rouge2_f += scores[i]['rouge-2']['f'] #ROUGE-2 F1 score

  # print(j, '{:5.4f}'.format(aux_rouge1/len(scores)))
  # print('{:5.4f}'.format(aux_rouge2/len(scores)))
  # print('{:5.4f}'.format(aux_fscore/len(scores)))

  aux2_rouge1_r.append(aux_rouge1_r/len(scores))
  aux2_rouge1_p.append(aux_rouge1_p/len(scores))
  aux2_rouge1_f.append(aux_rouge1_f/len(scores))
  aux2_rouge2_r.append(aux_rouge2_r/len(scores))
  aux2_rouge2_p.append(aux_rouge2_p/len(scores))
  aux2_rouge2_f.append(aux_rouge2_f/len(scores))

  aux_rouge1_r = 0
  aux_rouge1_p = 0
  aux_rouge1_f = 0
  aux_rouge2_r = 0
  aux_rouge2_p = 0
  aux_rouge2_f = 0

print('ROUGE-1 recall = {:5.4f}'.format(sum(aux2_rouge1_r)/len(foname_duc2002_sup)))
print('ROUGE-1 precision = {:5.4f}'.format(sum(aux2_rouge1_p)/len(foname_duc2002_sup)))
print('ROUGE-1 F1 score = {:5.4f}'.format(sum(aux2_rouge1_f)/len(foname_duc2002_sup)))
print('ROUGE-2 recall = {:5.4f}'.format(sum(aux2_rouge2_r)/len(foname_duc2002_sup)))
print('ROUGE-2 precision = {:5.4f}'.format(sum(aux2_rouge2_p)/len(foname_duc2002_sup)))
print('ROUGE-2 F1 score = {:5.4f}'.format(sum(aux2_rouge2_f)/len(foname_duc2002_sup)))

np.save(common_path + 'ROUGE1_recall_DUC2002A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge1_r)/len(foname_duc2002_sup))
np.save(common_path + 'ROUGE1_precision_DUC2002A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge1_p)/len(foname_duc2002_sup))
np.save(common_path + 'ROUGE1_F1score_DUC2002A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge1_f)/len(foname_duc2002_sup))
np.save(common_path + 'ROUGE2_recall_DUC2002A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge2_r)/len(foname_duc2002_sup))
np.save(common_path + 'ROUGE2_precision_DUC2002A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge2_p)/len(foname_duc2002_sup))
np.save(common_path + 'ROUGE2_F1score_DUC2002A_sup_cl{}'.format(num_cl) + '.npy', sum(aux2_rouge2_f)/len(foname_duc2002_sup))
```
